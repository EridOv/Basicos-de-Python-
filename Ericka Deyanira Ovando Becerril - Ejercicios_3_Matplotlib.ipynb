## Datos

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

importar matplotlib y agregar la linea para poder visualizar las gráficas en la consola de jupyter 

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

## Ejercicio 1

- Crear un objeto usando plt.figure()
- Agregar un  eje (axis) a la figura 
- Graficar los ejes x,y  
- Agregar las etiquetas correspondientes 


```python
fig=plt.figure()
axes=fig.add_axes([0.0,0.0,1.0,1.0])
axes.plot(x,y)
axes.set_xlabel("xlabel")
axes.set_ylabel("ylabel")
axes.set_title("title")
```

## Ejercicio 3

Crear un objeto  y agregar dos ejes (axis) , en las siguientes locaciones [0,0,1,1] y [0.2,0.5,0.2,0.2]  


```python
fig_a=plt.figure()
axes_a=fig_a.add_axes([0.0,0.0,1.0,1.0])
axes_b=fig_a.add_axes([0.2,0.5,0.2,0.2])
```

- Graficar x,y en ambos ejes y llamar a la figura  

```python
axes_a.plot(x,y)
axes_b.plot(x,y)
fig_a
```

## Ejercicio 2

Crear un objeto  y agregar dos ejes (axis) , en las siguientes locaciones [0,0,1,1] y [0.2,0.5,.4,.4]  


```python
fig_1=plt.figure()
axes_1=fig_1.add_axes([0.0,0.0,1.0,1.0])
axes_2=fig_1.add_axes([0.2,0.5,0.4,0.4])
```

Graficar x,y en el eje de menor tamaño  y x,z en eje principal, poner particular atención en los límites de las figuras 

```python
axes_1.plot(x,z)
axes_2.plot(x,y)
axes_2.set_xlim(20,22)
axes_2.set_ylim(30,50)
fig_1
```

## Ejercicio 4

Usar subplots de una fila dos columnas 

```python
fig_,axes_=plt.subplots(nrows=1,ncols=2)
```

Graficar (x,y) y (x,z) en los ejes (axis) con el grosor de 3 y el estilo adecuado 

```python
fig_z,axes_z=plt.subplots(nrows=1,ncols=2)
axes_z[0].plot(x,y, color='blue', ls="--",linewidth=3.0)
axes_z[1].plot(x,z, color='red', ls="-", linewidth=3.0)
```
